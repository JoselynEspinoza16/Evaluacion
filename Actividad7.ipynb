```python
# Solo columnas cuantitativas
cuantitativas = df1.select_dtypes(include=["int64", "float64"])

ax = cuantitativas.plot(kind="box", vert=False, figsize=(15, 8))
ax.set_title("Valores AtÃ­picos del Dataframe")
ax.set_xlabel("valor")
ax.set_ylabel("")       
plt.tight_layout()
plt.show()
```

```python
#MÃ©todo aplicando desviaciÃ³n estandar. Encuentro los valores extremos
y=cuantitativas
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
#df= cuantitativas[(y<=Limite_Superior)&(y>=Limite_Inferior)]
# Reemplazar outliers por el promedio de la columna
for col in cuantitativas.columns:
    promedio = cuantitativas[col].mean()
    df1[col] = np.where(
        (df1[col] > Limite_Superior[col]) | (df1[col] < Limite_Inferior[col]),
        promedio,
        df1[col]
    )

df1
```
